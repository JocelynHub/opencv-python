keras是深度学习中最容易使用的深度学习框架
tensorflow是谷歌做的
keras的底层是tensorflow

深度学习框架是对深度学习算法的封装
keras是对tensorflow做封装，使用更简单，但是效率也会变慢，慢两倍

```python
import keras
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential#一个模型的结构，按顺序构成的模型，比较简单 一层接一层
from keras.layers import Dense#全连接层
```

```python
#使用numpy生成100个随机点
x_data = np.random.rand(100)
noise = np.random.normal(0,0.01,x_data.shape)

y_data = x_data*0.1+0.2+noise

#显示随机点
plt.scatter(x_data,y_data)
plt.show()
```

```python
#构建顺序模型
model = Sequential()
#在模型中，添加一个全连接层
model.add(Dense(units=1,input_dim=1))#输入1维，输出1维
model.compile(optimizer="sgd",loss="mse")#随机梯度下降，均方误差

#训练
for step in range(3001):
    cost = model.train_on_batch(x_data,y_data)#每次训练一个批次
    if(step%500 ==0):
        print("cost:",cost)
#打印权值和偏置
W,b = model.layers[0].get_weights()
print("W:",W,"b:",b)

#x_data输入网络中得到预测值
y_pred = model.predict(x_data)
plt.scatter(x_data,y_data)
plt.plot(x_data,y_pred,'r-',lw=3)
plt.show()
```
